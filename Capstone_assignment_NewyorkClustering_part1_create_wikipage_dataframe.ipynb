### Import libraries

### Instantiate target dataframe

```python
column_names = ['Postal Code', 'Borough', 'Neighbourhood'] 
# instantiate the dataframe
tor_neighbourhoods = pd.DataFrame(columns=column_names)
```

### Extract wikipedia page using requests library

```python
wiki_page = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
html_doc =  requests.get(wiki_page)
print('Done')
```

### Use Beautiful soup to parse and store the extracted wikipage

```python
wikipage = BeautifulSoup(html_doc.text, 'html.parser') #read the extracted Wikipage using beautifulsoup
table = wikipage.find("table")
i=0
j=0
for row in table.findAll("tr"): 
    col = row.findAll("td")
    if len(col)>0:
        pcode = col[0].get_text().strip('\n')
        borough = col[1].get_text().strip('\n')
        neighbourhood = col[2].get_text().strip('\n')
        if borough != 'Not assigned':  #skips all boroughs which are having value = 'Not assigned'
            # if borough is legitimate ie having value other than Not assigned, store it to dataframe
            tor_neighbourhoods=tor_neighbourhoods.append({'Postal Code': pcode,
                                          'Borough': borough,
                                          'Neighbourhood': neighbourhood},ignore_index=True)
        else:
            j=j+1
        i=i+1
text = 'Skipped {} boroughs out of {} entries as they are <Not assigned>'.format(j,i)

#priting howmany entries skipped 
print(text)
```
